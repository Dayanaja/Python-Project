# Remove the outliers

```python
# Step 1: Remove outliers
Q1 = df_laptop['Price'].quantile(0.25)
Q3 = df_laptop['Price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Clean the dataset by filtering out the outliers
df_laptop_cleaned = df_laptop[(df_laptop['Price'] >= lower_bound) & (df_laptop['Price'] <= upper_bound)]

# Step 2: Visualize the comparison
plt.figure(figsize=(12, 6))

# Plot after removing outliers
plt.subplot(1, 2, 2)
sns.boxplot(x=df_laptop_cleaned['Price'])
plt.title('After Removing Outliers')

plt.tight_layout()
plt.show()
```

```python
# Step 1: Remove outliers for 'Weight'
Q1 = df_laptop['Weight'].quantile(0.25)
Q3 = df_laptop['Weight'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Clean the dataset by filtering out the outliers based on 'Weight'
df_laptop_cleaned = df_laptop[(df_laptop['Weight'] >= lower_bound) & (df_laptop['Weight'] <= upper_bound)]

# Step 2: Visualize the comparison
plt.figure(figsize=(12, 6))

# Plot after removing outliers
plt.subplot(1, 2, 2)
sns.boxplot(x=df_laptop_cleaned['Weight'])
plt.title('After Removing Outliers')

plt.tight_layout()
plt.show()
```

```python
# Step 1: Remove outliers for 'Inches'
Q1_inches = df_laptop['Inches'].quantile(0.25)
Q3_inches = df_laptop['Inches'].quantile(0.75)
IQR_inches = Q3_inches - Q1_inches

lower_bound_inches = Q1_inches - 1.5 * IQR_inches
upper_bound_inches = Q3_inches + 1.5 * IQR_inches

# Clean the dataset by filtering out the outliers based on 'Inches'
df_laptop_cleaned_inches = df_laptop[(df_laptop['Inches'] >= lower_bound_inches) & (df_laptop['Inches'] <= upper_bound_inches)]

# Step 2: Visualize the comparison for 'Inches'
plt.figure(figsize=(12, 6))

# Plot after removing outliers
plt.subplot(1, 2, 2)
sns.boxplot(x=df_laptop_cleaned_inches['Inches'])
plt.title('After Removing Outliers (Inches)')

plt.tight_layout()
plt.show()
```

# Hypothesis Testing for avg price of laptops

```python
from scipy.stats import ttest_1samp

# Assume we're testing if the mean price is â‚¹50,000
hypothesized_mean_price = 50000

# Perform a one-sample t-test
t_stat, p_val = ttest_1samp(df_laptop['Price'], hypothesized_mean_price)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")

if p_val < 0.05:
    print("Reject Hâ‚€: The mean price is significantly different from â‚¹50,000.")
else:
    print("Fail to Reject Hâ‚€: No significant difference from â‚¹50,000.")
```

# Hypothesis Testing for RAM & Price

```python
from scipy.stats import ttest_ind, ttest_1samp   # add ttest_ind here
import pandas as pd
import numpy as np

# Ensure RAM is numeric
df_laptop['Ram'] = pd.to_numeric(
    df_laptop['Ram'].astype(str).str.extract(r'(\d+)')[0],
    errors='coerce'
)

# Ensure Price is numeric
df_laptop['Price'] = pd.to_numeric(df_laptop['Price'], errors='coerce')

# Drop missing values
df_clean = df_laptop.dropna(subset=['Ram', 'Price'])

# Create two groups
group_low  = df_clean[df_clean['Ram'] <= 8]['Price']
group_high = df_clean[df_clean['Ram'] > 8]['Price']

# Perform twoâ€‘sample Welchâ€™s tâ€‘test
t_stat, p_val = ttest_ind(group_high, group_low, equal_var=False)

print(f"Tâ€‘statistic: {t_stat:.4f}")
print(f"Pâ€‘value:     {p_val:.4f}")

if p_val < 0.05:
    print("Reject Hâ‚€: Price differs significantly based on RAM size.")
else:
    print("Fail to Reject Hâ‚€: No significant price difference based on RAM.")
```

# Hypothesis Testing for chi-square on company and Typename

```python
import pandas as pd
from scipy.stats import chi2_contingency
import seaborn as sns

contingency_table = pd.crosstab(df_laptop['Company'], df_laptop['TypeName'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)

if p < 0.05:
    print("Reject the null hypothesis - variables are not independent")
else:
    print("Fail to reject the null hypothesis - variables are independent")
```
